```python
from sklearn.impute import SimpleImputer

# Imputar con la media
imputer = SimpleImputer(strategy='mean')
x_ent = imputer.fit_transform(x_ent)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection  import train_test_split # divide los datos para hacer el entrenamiento del modelo y el otro para probarlo
x_ent, x_pru, y_ent, y_pru = train_test_split(x, y, test_size=.2,)
```

```python
# Eliminar filas donde y_ent tiene NaN
valid_indices = y_ent.notnull()  # Obtener índices válidos
x_ent = x_ent[valid_indices]      # Filtrar x_ent
y_ent = y_ent[valid_indices]      # Filtrar y_en
```

```python
# Eliminar filas donde y_ent tiene NaN
valid_indices = y_pru.notnull()  # Obtener índices válidos
x_pru = x_pru[valid_indices]      # Filtrar x_ent
y_pru = y_pru[valid_indices]      # Filtrar y_ent
```

```python
model.fit(x_ent, y_ent)
```

```python
predic = model.predict(x_pru)
```

```python
comparacion = {"prediccion": predic, "Valor Real": y_pru}
pd.DataFrame(comparacion)
```

```python
print(model.score(x_ent, y_ent))
print(model.score(x_pru, y_pru))
```
